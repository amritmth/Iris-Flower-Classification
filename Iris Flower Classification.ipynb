## Correlation Matrix
A correlation matrix plays a fundamental role in the field of data analysis by revealing complex connections between variables within a dataset. It serves as a comprehensive tool for understanding how variables are interrelated, enabling the detection of patterns, dependencies, and valuable insights hidden within the data.

This matrix is essentially a tabular representation that displays the correlation coefficients between different variables present in the dataset. Correlation coefficients measure the strength and direction of the linear relationship between two variables. These coefficients generally fall within the range of -1 to 1, where:

A value of 1 indicates a perfect positive correlation, indicating that as one variable increases, the other variable also increases proportionally. A value of 0 indicates no correlation, which means there is no clear linear relationship between the variables. A value of -1 represents a perfect negative correlation, indicating that as one variable increases, the other variable decreases proportionally.

1). “PetalLengthCM” and “PetalWidthCM” have a very strong positive correlation of about 0.96,
          indicating that as petal length increases, petal width increases significantly.
     
2). “SepalLengthCm” and “SepalWidthCm” have a weak negative correlation of about -0.11, 
          meaning that as sepal length increases, sepal width may slightly decrease.

### Model Training

```python
X = iris.drop(columns=['Species'])
y = iris['Species']
```

### Split Data into Training and Testing Sets

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=1)
```

# Logistic Regression model

```python
from sklearn.linear_model import LogisticRegression
```

```python
# model fitting

Log=LogisticRegression()
Log.fit(X_train,y_train)
```

```python
print("Train score", Log.score(X_train, y_train))
print("Test score", Log.score(X_test, y_test))
```

```python
# model accuracy
print("Accuracy(Logistic Regression): ",Log.score(X_test,y_test)*100)
```

```python
# model accuracy
print("Accuracy(Logistic Regression): ",Log.score(X_test,y_test)*100)

# Your original accuracy value
accuracy = 0.9777777777777777

# Format the accuracy to display only two decimal places
formatted_accuracy = f'{accuracy:.2f}'

# model accuracy
print("Accuracy(Logistic Regression Overall Result): ",formatted_accuracy)
```

```python
Lr_pred=Log.predict(X_test)
```

```python
from sklearn.metrics import classification_report, confusion_matrix, f1_score,accuracy_score


# Confusion Matrix
cm = confusion_matrix(y_test, Lr_pred)
plt.figure(figsize=(8, 6))
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues')
plt.title('Confusion matrix of Logistic Regression')
plt.xlabel('Predicted')
plt.ylabel('True')
plt.show()

# Classification Report
print("Classification Report is:\n", classification_report(y_test, Lr_pred))


# Accuracy Score
print("Accuracy Score:", accuracy_score(y_test, Lr_pred))
```

### KFold Cross Validation

```python
from sklearn.model_selection import cross_val_score
from sklearn.model_selection import KFold
```

```python
kf = KFold(n_splits=5, shuffle=True, random_state=1)
```

```python
# Perform cross-validation on the model
scores = cross_val_score(Log, X, y, cv=kf)
```

```python
# Perform cross-validation and visualize the results
scores = cross_val_score(Log, X, y, cv=kf)
plt.figure(figsize=(10, 6))
plt.boxplot(scores)
plt.xlabel("Fold")
plt.ylabel("Score")
plt.title("Cross-Validation Scores")
plt.show()
```

```python
# Print the cross-validation scores
print("Cross-Validation Scores:", scores)
```

```python
# Calculate and print the mean score
mean_score = scores.mean()
print("Mean Score:", mean_score)
```
